```python
import requests
import urllib.request
import urllib3
import time
from bs4 import BeautifulSoup
import youtube_dl
import random
from pydub import AudioSegment
```

```python
type = 'eval' #balanced_train or eval 
label = 'speech' #label. Can be anything from Audioset (music, speech, vehicle, ...)
```

```python
from urllib.request import urlopen
import re

html = urlopen('https://storage.googleapis.com/audioset_website_data/youtube_corpus/v1/'+str(type)+'/'+str(label)+'/1.js')
bs = BeautifulSoup(html, 'lxml')
images = bs.find('p')
images = images.text

i = 0

end = len(images)
end = end - 35
ids = []
ids.append(images[5:16])

while i < end - 5:
    if images[i] == ']' and images[i+1] == ']' and images[i+2] ==']' and images[i+3] == ',' :
        ids.append(images[i+6:i+17])
        i = i+18
        
    i = i+1
    
```

```python
y = 0
while y < len(ids):

    link = 'https://www.youtube.com/watch?v=' + str(ids[y])

    y = y+1
    ydl_opts = {
        'ignoreerrors': True,
        'format': 'bestaudio/best',
        'outtmpl': '%(id)s.%(ext)s',

        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'wav',
            'preferredquality': '192',
        }],
    }

    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download([str(link)])
        info = ydl.extract_info(str(link))
        if info is not None:
            eyedn = info['id']
            
        if info is not None:    
            soundn = AudioSegment.from_wav(str(eyedn) + '.wav')
            soundn = soundn[1000:3000]
            soundn.export(str(eyedn) + '.wav', format="wav")
```
